#### AdaBoost and XGBoost
- Design an experiment to compare the performance of AdaBoost and XGBoost algorithms
- on a binary classification task using any dataset of your choice.
- Write Python code to implement both algorithms
- evaluate their performance using appropriate evaluation metrics.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./dataset/diabetes.csv')
df.sample(5)
```

```python
x = df.drop(['Outcome'], axis = 1)
x.head(2)
```

```python
y = df.drop(x.columns, axis = 1)
y.head(2)
```

```python
from sklearn.model_selection import train_test_split
# from sklearn.preprocessing import Binarizer
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 42, test_size = 0.3)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
```

```python
model1 = AdaBoostClassifier(random_state= 42)
model1.fit(x_train, y_train.to_numpy().ravel())
```

```python
model2 = XGBClassifier(random_state = 42)
model2.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
def el_plan(model, x_test, y_test):
    y_pred = model.predict(x_test)

    print("confusion matrix : ", confusion_matrix(y_test, y_pred))

    print("accuracy: ", accuracy_score(y_test, y_pred))
    print("precision: ", precision_score(y_test, y_pred))
    print("recall: ", recall_score(y_test, y_pred))
    print("f1: ", f1_score(y_test, y_pred))
    print("rac : ", roc_auc_score(y_test, y_pred))
```

```python
el_plan(model1, x_test, y_test)
```

```python
el_plan(model2, x_test, y_test)
```
