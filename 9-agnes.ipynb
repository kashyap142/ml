- Using any dataset implement Hierarchical Clustering (AGNES and DIANA).

- Plot the clustering output for the same dataset using these two hierarchical techniques.
- Compare the results.
- Write the inference.

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
df = load_iris()
x = df.data
y = df.target
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
plt.figure(figsize = (20, 6))
plt.title("AGNES Dendrogram")
plt.xlabel("Samples")
plt.ylabel("Distance")
dendrogram(linkage(x, method = 'ward'))
plt.show()
```

```python
plt.figure(figsize = (20, 6))
plt.title("DIANA Dendrogram")
plt.xlabel("Samples")
plt.ylabel("Distance")
dendrogram(linkage(x, method = 'single'))
plt.show()
```

```python
threshold = 2.5
agnes_cluster = fcluster(linkage(x, method = 'ward'), threshold, criterion = 'distance')
```

```python
plt.title("AGNES cluster")
plt.xlabel("Sepal length")
plt.ylabel("Sepal Width")
# plt.scatter(x[:, 0], x[:, 1], c=agnes_cluster, cmap='viridis')
plt.scatter(x[:, 0], x[:, 1], c=agnes_cluster)

plt.show()
```

```python
diana_cluster = fcluster(linkage(x, method='single'), threshold, criterion='distance')
```

```python
plt.title("DIANA cluster")
plt.xlabel("Sepal length")
plt.ylabel("Sepal Width")
# plt.scatter(x[:, 0], x[:, 1], c = diana_cluster, cmap = 'viridis')
plt.scatter(x[:, 0], x[:, 1], c = diana_cluster)

plt.show()
```
